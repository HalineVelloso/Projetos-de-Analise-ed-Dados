```python
import pandas as pd
import re
```

```python
df = pd.read_excel("ChavesClientes.xlsx", sheet_name = "base")
df
```

```python
df['Idade'] = df.ChaveSituacao.str[:2]
df['Genero'] = df.ChaveSituacao.str[2:3]
df['Estado Civil'] = df.ChaveSituacao.str[-1]
df
```

```python
columns = ['ChaveSituacao']
df.drop(columns, inplace=True, axis=1)
```

```python
df['Categoria'] = df.CatCliente.str.split('-').str.get(0)
df['VIP'] = df.CatCliente.str.split('-').str.get(1)
df
```

```python
columns = ['CatCliente']
df.drop(columns, inplace=True, axis=1)
```

```python
df['Risco'] = df.ClassRisco.apply(lambda x: re.findall('^[A-Z][^A-Za-z]?',x)[0])
df
```

```python
columns = ['ClassRisco']
df.drop(columns, inplace=True, axis=1)
df
```

```python
df.groupby(['Categoria', 'Risco']).agg({'Pagamento': [sum]})
```
